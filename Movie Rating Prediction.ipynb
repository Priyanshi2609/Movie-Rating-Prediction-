```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies_dataset=pd.read_csv("Movies_dataset.csv",encoding='ISO-8859-1')
```

```python
movies_dataset.head()
```

```python
movies_dataset.isnull().sum()
```

```python
movies_dataset=movies_dataset.dropna(subset=['Rating'])
```

```python
movies_dataset
```

```python
movies_dataset.isnull().sum()
```

```python
movies_dataset=movies_dataset.dropna(subset=['Actor 1','Actor 2','Actor 3','Director','Genre'])
```

```python
movies_dataset
```

```python
movies_dataset.isnull().sum()
```

```python
movies_dataset.info()
```

```python
movies_dataset.loc[:, 'Votes'] = movies_dataset['Votes'].str.replace(',', '', regex=False).astype(int)
```

```python
movies_dataset
```

```python
movies_dataset.loc[:, 'Duration'] = movies_dataset['Duration'].str.replace(' min', '', regex=False)
```

```python
movies_dataset.loc[:, 'Duration'] = pd.to_numeric(movies_dataset['Duration'])
movies_dataset.loc[:, 'Duration']=movies_dataset.fillna(movies_dataset['Duration'].median())
```

```python
movies_dataset.isnull().sum()
```

```python
movies_dataset.loc[:, 'Year']=movies_dataset['Year'].str.strip('()').astype(int)
```

```python
movies_dataset
```

```python
top_movies=movies_dataset.loc[movies_dataset['Rating'].sort_values(ascending=False)[:10].index]
```

```python
top_movies
```

```python
sns.histplot(data=top_movies,x='Year',hue='Rating',multiple='stack')
```

```python
genre_counts=movies_dataset['Genre'].value_counts().reset_index()
genre_counts.columns=['Genre','Count']
top_genre=genre_counts.head(5)
top_genre
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
trans_data=movies_dataset.drop(['Name'],axis=1)
trans_data['Director']=label_encoder.fit_transform(movies_dataset['Director'])
trans_data['Actor 1']=label_encoder.fit_transform(movies_dataset['Actor 1'])
trans_data['Actor 2']=label_encoder.fit_transform(movies_dataset['Actor 2'])
trans_data['Actor 3']=label_encoder.fit_transform(movies_dataset['Actor 3'])
trans_data['Genre']=label_encoder.fit_transform(movies_dataset['Genre'].apply(lambda x:', '.join(x)))
trans_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
sc_data=scaler.fit_transform(trans_data)
sc_df=pd.DataFrame(sc_data,columns=trans_data.columns)
sc_df.head()
```

```python
corr_df=trans_data.corr(numeric_only=True)
corr_df["Rating"].sort_values(ascending=False)
```

```python
sns.heatmap(corr_df,annot=False,cmap='coolwarm')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
x=trans_data.drop(['Rating'],axis=1)
```

```python
y=trans_data['Rating']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
x_test=np.array(x_test)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
print('R2 score: ',r2_score(y_test,y_pred))
print('Mean squared error: ',mean_squared_error(y_test,y_pred))
print('Mean absolute error: ',mean_absolute_error(y_test,y_pred))
```
